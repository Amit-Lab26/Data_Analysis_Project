```python
#Working with Date and Time Category
#Q1. what days of the month makes most sales?
sales['Date'] #check the dates dtype
```

```python
pd.to_datetime(sales['Date']) #date dtype is changed to datetime
```

```python
pd.to_datetime(sales['Date']).dt.day #we can get day,month, year acording to our needs
```

```python
sales.head(10)
```

```python
sales['Day']=pd.to_datetime(sales['Date']).dt.day #create a new column with name day
sales.head(5)
```

```python
print(sales.groupby('Day').sum()['Total'] ) #It combine the sales occured in a each month at day 1, day 2 ....
day_sales= sales.groupby('Day').sum()['Total'] #saving this in var day_sales
```

```python
day_sales.plot()
plt.show()
#we can see in the graph that people spends more on 15th and less on month end
```

```python
#which months make most sales 
sales['Months']=pd.to_datetime(sales['Date']).dt.month
```

```python
sales.head(5)
```

```python
print(sales.groupby('Months').sum()['Total'])
sales_by_month=sales.groupby('Months').sum()['Total']
#so Mar month has the max no. of sales
```

```python
sales_by_month.plot(kind='bar')
plt.grid() #to check exact which month has the max sales
plt.show()
```
